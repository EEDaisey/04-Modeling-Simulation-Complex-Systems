```python
# ###########################################################################
# Author: Edward E. Daisey
# Course: Modeling & Simulation of Complex Systems
# Title: Assignment 10 - Problem 7
# Date: 5th of April 2026
# ###########################################################################

################################### Overview ##################################
# Description:
#   This script solves Problem 7 from the assignment sheet:
#
#       x_(n+1) = r x_n (1 - x_n)
#
#   with
#
#       r = 1 + sqrt(z)
#       z = 11/3 + (2/3) * [ (100 + sqrt(9936))^(1/3) + (100 - sqrt(9936))^(1/3) ]
#
#   It produces:
#
#     (7a) High-precision values of z and r
#     (7b) x_1001, ..., x_1010 for several random initial conditions
#     (7c) A comparison with the orbit diagram
#
# Output:
#   Terminal output labeled by subsection.
# ###########################################################################

# ############################ Imports ######################################
from decimal import Decimal, getcontext
import random
# ###########################################################################


# ############################ Constants ####################################
getcontext().prec = 80

seedValue = 2026
numberOfInitialConditions = 4

tailStart = 1001
tailEnd = 1010
# ###########################################################################


# ############################ Function 1 ###################################
# Name:
#   DecimalCubeRoot
#
# Purpose:
#   Computes the real cube root of a Decimal using Newton iteration.
#
# Input:
#   value : Decimal number.
#
# Output:
#   Real cube root of value.
def DecimalCubeRoot(value):
    if value == 0:
        return Decimal(0)

    sign = Decimal(1)
    if value < 0:
        sign = Decimal(-1)
        value = -value

    currentGuess = Decimal(str(float(value) ** (1.0 / 3.0)))

    for _ in range(25):
        currentGuess = (2 * currentGuess + value / (currentGuess * currentGuess)) / 3

    return sign * currentGuess
# ###########################################################################


# ############################ Function 2 ###################################
# Name:
#   ComputeParameter
#
# Purpose:
#   Computes z and r to high precision.
#
# Input:
#   None.
#
# Output:
#   zValue : High-precision Decimal value of z.
#   rValue : High-precision Decimal value of r.
def ComputeParameter():
    sqrt9936 = Decimal(9936).sqrt()

    zValue = (
        Decimal(11) / Decimal(3)
        + (Decimal(2) / Decimal(3))
        * (DecimalCubeRoot(Decimal(100) + sqrt9936) + DecimalCubeRoot(Decimal(100) - sqrt9936))
    )

    rValue = Decimal(1) + zValue.sqrt()

    return zValue, rValue
# ###########################################################################


# ############################ Function 3 ###################################
# Name:
#   LogisticMap
#
# Purpose:
#   Evaluates one step of the logistic map.
#
# Input:
#   x      : Current state.
#   rValue : Logistic-map parameter.
#
# Output:
#   Next state.
def LogisticMap(x, rValue):
    return rValue * x * (Decimal(1) - x)
# ###########################################################################


# ############################ Function 4 ###################################
# Name:
#   GenerateInitialConditions
#
# Purpose:
#   Produces reproducible random initial conditions in (0, 1).
#
# Input:
#   None.
#
# Output:
#   List of Decimal initial conditions.
def GenerateInitialConditions():
    rng = random.Random(seedValue)
    initialConditions = []

    for _ in range(numberOfInitialConditions):
        initialConditions.append(Decimal(str(rng.random())))

    return initialConditions
# ###########################################################################


# ############################ Function 5 ###################################
# Name:
#   GenerateTail
#
# Purpose:
#   Computes x_1001 through x_1010 for a given initial condition.
#
# Input:
#   initialCondition : Starting value x0.
#   rValue           : Logistic-map parameter.
#
# Output:
#   List of (index, value) pairs.
def GenerateTail(initialCondition, rValue):
    currentValue = initialCondition
    tailValues = []

    for index in range(1, tailEnd + 1):
        currentValue = LogisticMap(currentValue, rValue)
        if index >= tailStart:
            tailValues.append((index, currentValue))

    return tailValues
# ###########################################################################


# ############################ Function 6 ###################################
# Name:
#   PrintTail
#
# Purpose:
#   Prints x_1001 through x_1010 for one initial condition.
#
# Input:
#   initialCondition : Starting value x0.
#   tailValues       : List of (index, value) pairs.
#
# Output:
#   Terminal output.
def PrintTail(initialCondition, tailValues):
    print(f"x0 = {initialCondition}")
    for index, value in tailValues:
        print(f"x_{index} = {value}")
    print()
# ###########################################################################


# ############################ Function 7 ###################################
# Name:
#   Main
#
# Purpose:
#   Executes parts (7a), (7b), and (7c) with clear subsection labels.
#
# Input:
#   None.
#
# Output:
#   Terminal output.
def Main():
    zValue, rValue = ComputeParameter()
    initialConditions = GenerateInitialConditions()

    print("Problem 7")
    print()

    print("(7a) High-precision parameter values")
    print(f"z = {zValue}")
    print(f"r = {rValue}")
    print()

    print("(7b) Values x_1001 through x_1010")
    for initialCondition in initialConditions:
        tailValues = GenerateTail(initialCondition, rValue)
        PrintTail(initialCondition, tailValues)

    print("(7c) Comparison with orbit diagram")
    print("All tested initial conditions approach the same stable period-3 orbit, up to cyclic order.")
    print("Therefore the results are consistent with the orbit diagram for the logistic map at this r.")
# ###########################################################################


# ############################# Execution ###################################
Main()
# ###########################################################################
```
